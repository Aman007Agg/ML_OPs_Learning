Importing Libraries and Loading Datasets

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Load datasets
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
tips = sns.load_dataset("tips")

# Display the first few rows of the datasets
print("Iris Dataset:")
print(iris.head(), "\n")

print("\nTitanic Dataset:")
print(titanic.head(), "\n")

print("\nTips Dataset:")
print(tips.head())

```

Line Plot

```python
# Line plot for Iris dataset
plt.figure(figsize=(10, 6))
plt.plot(iris.index, iris['sepal_length'], label='Sepal Length')
plt.plot(iris.index, iris['petal_length'], label='Petal Length', linestyle='--')
plt.title("Line Plot of Sepal and Petal Length")
plt.xlabel("Index")
plt.ylabel("Length (cm)")
plt.legend()
plt.show()
```

Bar plot

```python
# Bar Plot for Titanic survival count
survival_count = titanic['survived'].value_counts()
plt.figure(figsize=(6, 4))
plt.bar(survival_count.index, survival_count.values, color=['red', 'green'])
plt.title("Bar Plot of Titanic Survival Count")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.xticks([0, 1], ['Not Survived', 'Survived'])
plt.show()
```

Histogram

```python
# Histogram for age distribution
plt.figure(figsize=(8, 5))
plt.hist(titanic['age'].dropna(), bins=10, color='skyblue', edgecolor='black')
plt.title("Age Distribution Histogram")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Scatter Plot

```python
# Scatter plot for Iris Dataset
plt.figure(figsize=(10, 6))
plt.scatter(iris['sepal_length'], iris['petal_length'], color='purple', alpha=0.7)
plt.title("Scatter Plot of Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
```

Pair Plot


```python
# Pair Plot for Iris Dataset
sns.pairplot(iris, hue='species', palette='dark')
plt.suptitle("Pair Plot of Iris Dataset")
plt.show()
```

Box Plot

```python
# Box Plot for Titanic ange by class with hue
plt.figure(figsize=(8, 6))
sns.boxplot(x='class', y='age', data=titanic, hue='class', palette='pastel', legend=False)
plt.title("Box Plot of Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")
plt.show()
```

Count Plot

```python
# Count plot for Titanic survival by Gender
plt.figure(figsize=(6, 4))
sns.countplot(x='sex', hue='survived', data=titanic, palette='coolwarm')
plt.title("Count Plot of Titanic Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

Heatmap

```python
# Select numeric columns for correlation
numeric_titanic = titanic.select_dtypes(include=['float64', 'int64'])

# Calculate correlation matrix
correlation_matrix = numeric_titanic.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()


```

Time Series Plot


```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: sales over months
data = {
    'Date': pd.date_range(start='2020-01-01', periods=12, freq='ME'),
    'Sales': [200, 210, 215, 220, 230, 250, 245, 260, 270, 280, 290, 300]
}

# Create a Dataframe
df = pd.DataFrame(data)

# Set the 'Date' column as the index
df.set_index('Date', inplace=True)

# Plot the time series data
print(df)
```

```python
# Potting the time series data
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Sales'], marker='o', color='b',linestyle='-', label='Sales')
plt.title('Monthly Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure()

# Add the sales data to the figure
fig.add_trace(go.Scatter(x=df.index, y=df['Sales'], mode='lines+markers', name='Sales'))

# Customize the layout
fig.update_layout(
    title='Monthly Sales Over Time',
    xaxis_title='Month',
    yaxis_title='Sales',
    showlegend=True,
    template='plotly_dark',
    hovermode='x unified',
    hoverlabel=dict(
        bgcolor='white',
        bordercolor='black',
        font=dict(color='black')
    ),
)

# Show the figure
fig.show()
```

```python
import plotly.express as px
# Plotly express example
fig = px.line(df, x=df.index, y='Sales', title='Monthly Sales Over Time')
fig.update_traces(mode='lines+markers')
fig.show()
```
