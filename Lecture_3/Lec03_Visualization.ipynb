Importing Libraries and Loading Datasets

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load datasets
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")

# Display the first few rows of the datasets
print("Iris Dataset:")
print(iris.head(), "\n")

print("Titanic Dataset:")
print(titanic.head(), "\n")
```

Line Plot

```python
# Line plot for Iris dataset
plt.figure(figsize=(8, 5))
plt.plot(iris.index, iris['sepal_length'], label='Sepal Length')
plt.plot(iris.index, iris['petal_length'], label='Petal Length', linestyle='--')
plt.title('Line Plot of Sepal and Petal Length')
plt.xlabel('Index')
plt.ylabel('Length (cm)')
plt.legend()
plt.show()
```

Bar Plot


```python
# Bar plot for Titanic survival count
survival_counts = titanic['survived'].value_counts()
plt.figure(figsize=(6, 4))
plt.bar(survival_counts.index, survival_counts.values, color=['red', 'green'])
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

Histogram

```python
# Histogram for age distribution
plt.figure(figsize=(8, 5))
plt.hist(titanic['age'].dropna(), bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Scatter Plot

```python
# Scatter plot for Iris dataset
plt.figure(figsize=(7, 5))
plt.scatter(iris['sepal_length'], iris['petal_length'], color='purple', alpha=0.7)
plt.title('Scatter Plot of Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```

Pair Plot

```python
# Pair plot for Iris dataset
sns.pairplot(iris, hue='species', palette='dark')
plt.show()
```

Box Plot

```python
# Box plot for Titanic age by class with hue
plt.figure(figsize=(8, 6))
sns.boxplot(data=titanic, x='class', y='age', hue='class', palette='Set2', legend=False)
plt.title('Box Plot of Age by Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

Count Plot

```python
# Count plot for Titanic survival by Gender
plt.figure(figsize=(7, 5))
sns.countplot(data=titanic, x='sex', hue='survived', palette='pastel')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

Heatmap

```python
# Select numeric columns for correlation
numeric_titanic = titanic.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
correlation = numeric_titanic.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Correlation Heatmap of Titanic Dataset')
plt.show()
```

Time Series Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data: sales over months
data = {
    'Date': pd.date_range(start='2020-01-01', periods=12, freq='ME'),  # 12 months starting from Jan 2020
    'Sales': [200, 210, 215, 220, 230, 250, 245, 260, 270, 280, 290, 300]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the first few rows of the dataset
print(df)
```

```python
# Plotting the time series data
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Sales'], marker='o', color='b', label='Sales')
plt.title('Monthly Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure()

# Add the sales data to the figure
fig.add_trace(go.Scatter(x=df['Date'], y=df['Sales'], mode='lines+markers', name='Sales'))

# Customize layout
fig.update_layout(
    title='Monthly Sales Over Time',
    xaxis_title='Date',
    yaxis_title='Sales',
    template='plotly_dark',
    hovermode='x unified',
)

# Show the plot
fig.show()
```

```python
import plotly.express as px

# Plotly express example
fig = px.line(df, x='Date', y='Sales', title='Monthly Sales Over Time')
fig.update_traces(mode='lines+markers')
fig.show()
```
